```python
# function that counts the number of monograms
def count_monograms(dict):   
    
    input = open(input_encrypted_path, 'r')
    
    for line in input:
        line = re.sub(r'[^а-яёЁА-Я]+', '', line).lower()
        
        for i in range(0, len(line)):
            for j in range(1, 2):
                ch = line[i:i+j] # substring with monogram

                if ch in dict:
                    dict[ch] += 1
                else:
                    dict[ch] = 1
                    
    return dict


# function that counts the number of monograms
def count_bigrams(dict):   
    
    input = open(input_encrypted_path, 'r')
    
    for line in input:
        line = re.sub(r'[^а-яёЁА-Я]+', '', line).lower()
        
        for i in range(0, len(line)):
            for j in range(2, 3):
                bi = line[i:i+j] # substring with bigram

                if bi in dict:
                    dict[bi] += 1
                else:
                    dict[bi] = 1
                    
    return dict


# function that counts the number of trigrams
def count_trigrams(dict):   
    
    input = open(input_encrypted_path, 'r')
    
    for line in input:
        line = re.sub(r'[^а-яёЁА-Я]+', '', line).lower()
        
        for i in range(0, len(line)):
            for j in range(3, 4):
                tri = line[i:i+j] # substring with trigram

                if tri in dict:
                    dict[tri] += 1
                else:
                    dict[tri] = 1
                    
    return dict


# function that sorts dictionaries by frequencies
def sort_by_frequency(dict):
    return sorted(dict.items(), key=lambda x: x[1], reverse=True)


# function that checks the size of the dictionaries and clips the larger one to the smaller one
def trim_dictionaries(sample_frequencies, encrypted_frequencies):
    
    len_sample = len(sample_frequencies)
    len_encrypted = len(encrypted_frequencies)
    
    if len_sample > len_encrypted:
        sample_frequencies = sample_frequencies[:len_encrypted]
    elif len_sample < len_encrypted:
        encrypted_frequencies = encrypted_frequencies[:len_sample]
    
    return sample_frequencies, encrypted_frequencies


# function that maps frequencies and creates a dictionary for decryption
def create_decryption_dictionary(sample_frequencies_dict, encrypted_frequencies_dict):
    
    decryption_dict = {}
    for i in range(0, len(encrypted_frequencies_dict)):
        decryption_dict[encrypted_frequencies_dict[i][0]] = sample_frequencies_dict[i][0]

    return decryption_dict
        
        
# function that replaces all mono-/bi-/trigrams
def decrypt(text):

    decrypted_text = []
            
    # for each text character replace the character with a dict value 
    # if the character is a dict key corresponding to this value
    for ch in text:
        for key, value in decryption_dict_mon.items():
            if ch == key:
                decrypted_text.append(value)     
    text = ''.join(decrypted_text) # makes a single line out of this
    
    for key, value in decryption_dict_tri.items():
        text = text.replace(key, value)
    
    for key, value in decryption_dict_bi.items():
        text = text.replace(key, value)
    
    return text


# helper function that replaces a monogram
def replaceMonogram(text):
    
    mon_to_replace = str(input('Enter the monogram to replace: ')).lower()
    mon_new = str(input('Enter a new monogram: ')).lower()
    if len(mon_to_replace) == 1 and mon_to_replace.isalpha():
        text = text.replace(mon_to_replace, mon_new)
    else:
        print('Not a monogram!')

    return text   
```
